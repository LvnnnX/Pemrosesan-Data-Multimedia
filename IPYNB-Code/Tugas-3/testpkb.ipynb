```python
import pandas as pd
import numpy as np
```

```python
teks = ['data science is one of the most important fields of science',
          'this is one of the best data science courses',
          'data scientists analyze data']
```

```python
all_unique = np.array([],dtype=str)
for words in teks:
    all_unique = np.append(all_unique, np.unique(words.split()))
all_unique = np.unique(all_unique)
```

```python
df = pd.DataFrame(columns=all_unique)
```

```python
df
```

```python
for num,value in enumerate(teks):
    list_value = np.array([],dtype=int)
    for num_inside, value_inside in enumerate(value.split()):
        list_value = np.append(list_value, value.count(value_inside))
        # print(value_inside)
        df.loc[f'{num}',f'{value_inside}'] = list_value[num_inside]
```

```python
df.index.name = 'Docs'
df = df.apply(lambda x: x.fillna(0),axis=1)
df
```

```python
@lambda _:_(df)
def tf(df:pd.DataFrame=None)->pd.DataFrame:
    df = df.apply(lambda x: x / x.sum(),axis=1)
    display(df)
    return df
```

```python
@lambda _:_(df)
def tf_idf(df:pd.DataFrame=None)->pd.DataFrame:
    df = df.apply(lambda x: x * np.log(len(df.columns)/x.sum()),axis=1)
    display(df)
    return df
```
